```python
import cv2
import numpy as np
```

```python
img = cv2.imread('Cat_image.jpg') #method to read image.
```

```python
img.shape
```

```python
# Resize the image
img = cv2.resize(img, (800, 500)) #method to resize the image.
```

```python
type(img) #numpy array
```

```python
cv2.imshow('Cat Image', img) #method to display image.
cv2.waitKey(0) #waits for a key press to close the image window.
```

```python
#method to convert color image to grayscale. The second argument specifies the color conversion code.
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
```

```python
gray_img.shape
```

```python
# Display the grayscale image
cv2.imshow('Cat Image', gray_img)
cv2.waitKey(0)
```

```python
gray_img.shape #grayscale image has only one channel, so the shape is (height, width) instead of (height, width, channels)
```

```python
img.shape
```

```python
#let's resize the image.
img_resize = cv2.resize(img, (500, 500), interpolation=cv2.INTER_LINEAR) #method to resize image. The second argument is a tuple specifying the new size (width, height).
```

```python
img_resize.shape
```

```python
# Display image
cv2.imshow('Cat Image', img_resize)
cv2.waitKey(0)
```

```python
#if you don't know the image dimensions and still want to resize
#it to a certiain percentage of the original size, you can do the following:
img_resize1 = cv2.resize(img, (img.shape[1]//2, img.shape[0]//2)) #resize to half the original size. The new width is half of the original width (img.shape[1]//2) and the new height is half of the original height (img.shape[0]//2).
```

```python
img_resize1.shape
```

```python
#flipping the image.
img_flip = cv2.flip(img, 0) #method to flip image. The second argument specifies the flip code. 1 means horizontal flip, 0 means vertical flip, and -1 means both horizontal and vertical flip.
```

```python
cv2.imshow('flipped image', img_flip)
cv2.waitKey(0)
```

```python
#cropping image 
img_crop = img[100:400, 100:400] #method to crop image. The first set of indices (100:400) specifies the range of rows to keep, and the second set of indices (100:400) specifies the range of columns to keep.
```

```python
cv2.imshow('Cropped image', img_crop)
cv2.waitKey(0)
```

```python
cv2.imwrite('Saved_image.png', img_crop) #method to save image. The first argument is the filename, and the second argument is the image to be saved.
```

```python
# Drawing shapes and text on the image
img_d = np.zeros((500, 500, 3), dtype=np.uint8) #create a blank image of size 500x500 pixels with 3 color channels (BGR) and data type uint8 (unsigned 8-bit integer).
cv2.imshow('Blank Image', img_d)
cv2.waitKey(0)
```

```python
#Rectangle
cv2.rectangle(img_d, pt1=(100,100), pt2=(300,300), color=(255,0,0), thickness=5) #method to draw a rectangle on the image. The first argument is the image on which to draw, pt1 and pt2 are the top-left and bottom-right corners of the rectangle, color is the color of the rectangle in BGR format, and thickness is the thickness of the rectangle border.
#circle
cv2.circle(img_d, center=(250,250), radius=50, color=(0,255,0), thickness=5) #method to draw a circle on the image. The first argument is the image on which to draw, center is the center of the circle, radius is the radius of the circle, color is the color of the circle in BGR format, and thickness is the thickness of the circle border.
#line
cv2.line(img_d, pt1= (0,0), pt2=(512,512), color=(0,0,255), thickness=5) #method to draw a line on the image. The first argument is the image on which to draw, pt1 and pt2 are the starting and ending points of the line, color is the color of the line in BGR format, and thickness is the thickness of the line.
#text 
cv2.putText(img_d, text='OpenCV', org=(100,100), fontScale=4, color=(255,255,255), lineType=cv2.LINE_AA, thickness=2, fontFace=cv2.FONT_HERSHEY_SIMPLEX) #method to put text on the image. The first argument is the image on which to put the text, text is the string of text to be displayed, org is the bottom-left corner of the text in the image, fontFace is the font type, fontScale is the scale factor for the font size, color is the color of the text in BGR format, and thickness is the thickness of the text.
cv2.imshow('Image Drawings', img_d)
cv2.waitKey(0)
```
